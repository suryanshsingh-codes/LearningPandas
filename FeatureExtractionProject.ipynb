**Make a new coloumn separately for Episode Count: here we will use feature extraction**

```python
#for this i will create a function
def extract_episodes(txt): # this text is the "title" of the index numbers
    check=False
    data='' # here i will extract the data of count of episodes
    
    for i in txt:
        if i==')':
            break # if bracket is closed, break it
        
        if check==True:
            data+=i # extract when the check is true
            
        if i=='(':
            check=True #if bracket is open, continue for next step
    return data
# now i will apply this function on Each title of my data  AND from each title, i will extract count of episoded          
    
```

```python
#create a new coloumn
df["Episodes"]=df["Title"].apply(extract_episodes)
df.head()
```

```python
# now converting the count of episodes into numerical data
df['Episodes']= df['Episodes'].str.replace(" eps","")
#now convert these episodes into integer because these are string
df["Episodes"]=df["Episodes"].astype(int)
df.head()
```

**make a new column for time stamp**

```python
df.loc[2]['Title'] # watch closely the time stamp : "Apr 2015 - Mar 2016" : total  19 is its length
```

```python
def extraction_time(txt):
    check = False
    data = "" # here i will extract the data of count of episodes
    for i in range(len(txt)):
        if txt[i] == ')': # time stamp start after ")"
            for j in range(i+1,i+20): # start extracting data 20 steps more from one step forward to the pointer
                data += txt[j]

            return data
df.head()
```

```python
# create a new coloumn and apply above function
df['Total Time'] = df['Title'].apply(extraction_time)
df.head()
```

**create a new coloumn of months count**

```python
# Import relativedelta to calculate the difference between two dates in years and months
from dateutil.relativedelta import relativedelta

# Import datetime to convert string dates into datetime objects
from datetime import datetime


# Function to calculate the total number of months from a given time period
def calculate_total_months(period):
    try:
        # Split the string into start and end dates
        # Example: "Jan 2018 - Mar 2020"
        # start_str = "Jan 2018"
        # end_str = "Mar 2020"
        start_str, end_str = period.split(' - ')

        # Convert the start and end date strings into datetime objects
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')

        # Calculate the difference between the two dates
        r = relativedelta(end_date, start_date)

        # Convert years into months and add the remaining months
        # +1 is added to include the starting month
        # Example:
        # Jan 2020 to Mar 2020
        # Difference = 2 months
        # Including Jan = 3 months
        total_months = r.years * 12 + r.months + 1

        return total_months

    # If the date format is invalid or missing, return None
    except:
        return None


# Apply the function to every value in the 'Total Time' column
# and store the result in a new column called 'Months'
df['Months'] = df['Total Time'].apply(calculate_total_months)
df.head()
```
