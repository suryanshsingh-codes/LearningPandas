# **Missing Data**

**Finding The Missing Data**

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna() #find where are missing data
```

```python
df.isna().sum() #find the sum ciunt of missing data
```

```python
df.isna().any() #find whether there is any missing data in anny row
```

**Removing Null Values**

```python
df.dropna() #drop the rows containing missing data
```

```python
df.dropna(thresh=3) # drop the rows which are containing min 3 null values
```

**Filling Null Values**

```python
df.fillna(433) # fill all null values with 433
```

```python
df.fillna("I was a Null Value") 
```

```python
values = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```

# **Merging, Joining and Concatenating two DataFrames**

**Merging 2 DataFrames**

```python
# DataFrame 2: Emplyoee information
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries) # merge the two dataframes BUT only which are common in both
```

```python
pd.merge(employees,salaries,on='employee_id') # merge two dataframes, BUT which are common, and common should be employee_id
```

```python
pd.merge(employees,salaries,on='employee_id',how ='inner') 
# merge two dataframes, BUT which are common, and common should be employee_id, BUT "OUTER" means include all the values which are only common.
```

```python
pd.merge(employees,salaries,on='employee_id',how ='outer') 
# merge two dataframes, BUT which are common, and common should be employee_id, BUT "OUTER" means include all the values which are not common also
```

```python
pd.merge(employees,salaries,on='employee_id',how ='left') 
#merge two data frames, common should be merged, search common on basis of employee, 
# "left" means include all the emplyoee id which are in emplyoe_id
# because ( emplyoees, salaries), employee on the left
```

```python
pd.merge(employees,salaries,on='employee_id',how ='right')
```

# **GroupBy, Aggregation**

**Group by Category and calculate the sum of Sales**

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
df
```

```python
cat = df.groupby('Category')['Sales'].sum() # calculate sum of sales into all the categories
cat
```

**Group by multiple columns :**
Group by Category and Store


```python
cat = df.groupby(['Category','Store'])['Sales'].sum() # calculate sum of sales into all the category and into all the stores
cat
```

**Aggregation**

```python
""" only one aggregate function """
df['Sales'].mean()
```

```python
df['Sales'].agg(['sum', 'mean', 'min', 'max', 'count', 'std', 'median'])
```

# **Pivot Tables and Cross Tabs**

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)
df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df, values=['Sales', 'Units'], index='Region', columns='Product', aggfunc='median') 
#create pivot table from dataframe "df"
# but include only 2 values for pivot tables= "sales", "units" AND values must  be numeric
# the values should have for all the "reigions"->index,
# coloumns should be "product"
# 
# this aggregation function is optional, by defualt it's mean, # HERE WE USE AGGREGATE FUCNTIONS


# sales and unit values for different products different regions
```

```python
pd.crosstab(df['Region'],df['Product'])
# cross tab from the dataframe's -> Region coloumn and Product coloumn
# HERE WE USE COUNTING FUCNTIONS
```
